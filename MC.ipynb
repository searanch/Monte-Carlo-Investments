Monte Carlo Model to model potential investment returns.
Assumptions: Based on 20 years of an all stock portfolio. Using historical return of SP 500. Starting investment of $10,000 with 21k per year in addtional contributions. 

```julia
# I've created a chain of functions to simulate potential financial returns. 

function y1()
    
    # Generate random data based on an average return, standard deviation, min and max return
    random = rand(Truncated(Normal(.077, .19), -.47, .46), 10000) 
    
    # reshape data for a dataframe
    a = reshape(random, length(random),1)
    df = DataFrame(a,["y1return"]) # make a dataframe with each row being a potential return
    
    # Create two additional columns in the DataFrame
    df2 = DataFrame(invest=repeat(10000:10000, inner=10000))# run the simulations
    df3 = DataFrame(y1addfunds=repeat(21000:21000,inner=10000)) # add 
    
    # merge the data frames
    merged = hcat(df, df2,df3)
    merged[!,:y1total] = merged[!,:y1return] .* merged[!,:invest] + merged[!,:invest] + merged[!, :y1addfunds]
    merged
end

y1()

# Define function y2
function y2()
    # Generate random data
    random = rand(Truncated(Normal(.077, .19), -.47, .46), 10000)
    
    # Reshape data into a DataFrame
    a = reshape(random, length(random),1)
    df = DataFrame(a,["y2return"])
    
    # Call y1() function to obtain merged DataFrame
    merged = y1()
    
    # Create additional column in the DataFrame
    df2 = DataFrame(y2addfunds=repeat(21000:21000,inner=10000))
    
    # Merge the DataFrames
    merged2 = hcat(merged,df,df2)
    
    # Calculate the total value for each row in the DataFrame
    merged2[!,:y2total] = merged2[!,:y2return] .* merged2[!,:y1total] + merged2[!,:y1total] + merged2[!, :y2addfunds]
    
    # Return the merged DataFrame
    merged2
end

# define function y3()
function y3()
    # Generate Random Data
    random = rand(Truncated(Normal(.077, .19), -.47, .46), 10000)
    
   
    # Call y2() function to obtain merged DataFrame
    a = reshape(random, length(random),1)
    df = DataFrame(a,["y3return"])
    
    
    # Create two additional columns in the DataFrame
    merged = y2() 
    df2 = DataFrame(y3addfunds=repeat(21000:21000,inner=10000))
    merged2 = hcat(merged,df,df2)
    
    
    # merge the data frames
    merged2[!,:y3total] = merged2[!,:y3return] .* merged2[!,:y2total] + merged2[!,:y2total] + merged2[!, :y3addfunds]
   
    merged2
end

#TODO add comments to the rest of the function
function y4()
    random = rand(Truncated(Normal(.077, .19), -.47, .46), 10000)
    a = reshape(random, length(random),1)
    df = DataFrame(a,["y4return"])

    merged = y3()
    df2 = DataFrame(y4addfunds=repeat(21000:21000,inner=10000))
    merged2 = hcat(merged,df,df2)
    
    merged2[!,:y4total] = merged2[!,:y4return] .* merged2[!,:y3total] + merged2[!,:y3total] + merged2[!, :y4addfunds]
    # merged
    merged2
end
y4()

function y5()
    random = rand(Truncated(Normal(.077, .19), -.47, .46), 10000)
    a = reshape(random, length(random),1)
    df = DataFrame(a,["y5return"])

    merged = y4()
    df2 = DataFrame(y5addfunds=repeat(21000:21000,inner=10000))
    merged2 = hcat(merged,df,df2)
    
    merged2[!,:y5total] = merged2[!,:y5return] .* merged2[!,:y4total] + merged2[!,:y4total] + merged2[!, :y5addfunds]
    # merged
    merged2
end

y5()


function y6()
    random = rand(Truncated(Normal(.077, .19), -.47, .46), 10000)
    a = reshape(random, length(random),1)
    df = DataFrame(a,["y6return"])

    merged = y5()
    df2 = DataFrame(y6addfunds=repeat(21000:21000,inner=10000))
    merged2 = hcat(merged,df,df2)
    
    merged2[!,:y6total] = merged2[!,:y6return] .* merged2[!,:y5total] + merged2[!,:y5total] + merged2[!, :y6addfunds]
    # merged
    merged2
end

y6()

function y7()
    random = rand(Truncated(Normal(.077, .19), -.47, .46), 10000)
    a = reshape(random, length(random),1)
    df = DataFrame(a,["y7return"])

    merged = y6()
    df2 = DataFrame(y7addfunds=repeat(21000:21000,inner=10000))
    merged2 = hcat(merged,df,df2)
    
    merged2[!,:y7total] = merged2[!,:y7return] .* merged2[!,:y6total] + merged2[!,:y6total] + merged2[!, :y7addfunds]
    # merged
    merged2
end

y7()

function y8()
    random = rand(Truncated(Normal(.077, .19), -.47, .46), 10000)
    a = reshape(random, length(random),1)
    df = DataFrame(a,["y8return"])

    merged = y7()
    df2 = DataFrame(y8addfunds=repeat(21000:21000,inner=10000))
    merged2 = hcat(merged,df,df2)
    
    merged2[!,:y8total] = merged2[!,:y8return] .* merged2[!,:y7total] + merged2[!,:y7total] + merged2[!, :y8addfunds]
    # merged
    merged2
end

y8()

function y9()
    random = rand(Truncated(Normal(.077, .19), -.47, .46), 10000)
    a = reshape(random, length(random),1)
    df = DataFrame(a,["y9return"])

    merged = y8()
    df2 = DataFrame(y9addfunds=repeat(21000:21000,inner=10000))
    merged2 = hcat(merged,df,df2)
    
    merged2[!,:y9total] = merged2[!,:y9return] .* merged2[!,:y8total] + merged2[!,:y8total] + merged2[!, :y9addfunds]
    # merged
    merged2
end

y9()

function y10()
    random = rand(Truncated(Normal(.077, .19), -.47, .46), 10000)
    a = reshape(random, length(random),1)
    df = DataFrame(a,["y10return"])

    merged = y9()
    df2 = DataFrame(y10addfunds=repeat(21000:21000,inner=10000))
    merged2 = hcat(merged,df,df2)
    
    merged2[!,:y10total] = merged2[!,:y10return] .* merged2[!,:y9total] + merged2[!,:y9total] + merged2[!, :y10addfunds]
    # merged
    merged2
end

y10()


```
